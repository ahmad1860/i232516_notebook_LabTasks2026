```python
## TASK 2


routes = {
    "W": {"distance": 300, "time": 25, "congestion": "Medium"},
    "X": {"distance": 200, "time": 15, "congestion": "Low"},
    "Y": {"distance": 150, "time": 20, "congestion": "Medium"},
    "Z": {"distance": 250, "time": 10, "congestion": "High"}
}

congestion_weight = {
    "Low": 3,
    "Medium": 2,
    "High": 1
}

def calculate_utility(route):
    d = route["distance"]
    t = route["time"]
    c = congestion_weight[route["congestion"]]

    return -(d * 2 + t * 3.5 + c)

scores = {}

for name in routes:
    scores[name] = calculate_utility(routes[name])

optimal_route = max(scores, key=lambda k: scores[k])

for r, val in scores.items():
    print(f"Path {r} : {val}")

print(f"\nSelected Path: {optimal_route} with utility {scores[optimal_route]}")
```

```python
## TASK 3 
garden = {
    "Section 1": "Healthy",
    "Section 2": "Wilted"
}

beliefs = {
    "Section 1": "Unknown",
    "Section 2": "Unknown"
}

current_position = "Section 1"

def sense(area):
    return garden[area]

def irrigate(area):
    garden[area] = "Healthy"

def switch_area(area):
    return "Section 2" if area == "Section 1" else "Section 1"

steps = 6

for step in range(steps):
    print(f"\n--- Time Step {step + 1} ---")
    print(f"Agent location: {current_position}")

    observed_state = sense(current_position)
    beliefs[current_position] = observed_state

    if observed_state == "Wilted":
        action_taken = "Water"
        irrigate(current_position)
    else:
        action_taken = "Move"
        current_position = switch_area(current_position)

    print(f"Action taken: {action_taken}")
    print("Agent internal state:", beliefs)
    print("Actual environment:", garden)


```
