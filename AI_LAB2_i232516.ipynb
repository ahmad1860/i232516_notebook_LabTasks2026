## task 7

```python
df_tips = sns.load_dataset("tips")
Q1_tb = df_tips["total_bill"].quantile(0.25)
Q3_tb = df_tips["total_bill"].quantile(0.75)
IQR_tb = Q3_tb - Q1_tb

lower_tb = Q1_tb - 1.5 * IQR_tb
upper_tb = Q3_tb + 1.5 * IQR_tb
outliers_total_bill = df_tips[(df_tips["total_bill"] < lower_tb) |(df_tips["total_bill"] > upper_tb)]
Q1_tip = df_tips["tip"].quantile(0.25)
Q3_tip = df_tips["tip"].quantile(0.75)
IQR_tip = Q3_tip - Q1_tip
lower_tip = Q1_tip - 1.5 * IQR_tip
upper_tip = Q3_tip + 1.5 * IQR_tip
outliers_tip = df_tips[ (df_tips["tip"] < lower_tip) | (df_tips["tip"] > upper_tip)]

print("Total_bill ", outliers_total_bill.shape[0])
print("Tip outlier", outliers_tip.shape[0])

plt.boxplot(df_tips["total_bill"])
plt.show()
plt.boxplot(df_tips["tip"])
plt.show()




```
